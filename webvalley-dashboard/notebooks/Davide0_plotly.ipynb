```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('data/merged_APPA_data.csv')
df.Data = pd.to_datetime(df.Data)
```

```python
stations = ['Monte Gaza', 'Riva del Garda', 'Parco S. Chiara', 'Piana Rotaliana', 'Rovereto', 'Via Bolzano', 'A22 (Avio)', 'Borgo Valsugana']
pollutants = ['Ozono', 'Biossido di Azoto', 'PM2.5', 'PM10', 'Ossido di Carbonio', 'Biossido Zolfo']
threshold = {'PM10': [36, 50], 
             'Biossido di Azoto': [101, 200], 
             'Ozono': [121, 180], 
             'PM2.5': [201, 350], 
             'Ossido di Carbonio': [7.6, 10], 
             'Biossido Zolfo': [201, 350]}
```

```python
def subdf(df, pollutant, station):
    mask = (df.Inquinante == pollutant) & (df.Stazione == station)
    subdf = df[mask]

    return subdf

def resample_data(df, resample_rule):
    df_mean = df.resample(resample_rule, on='Data').mean()
    df_mean = df_mean.reset_index()
    return df_mean
```

```python
pm10_a22 = subdf(df, pollutants[3], stations[6])
pm10_a22_resampled = resample_data(pm10_a22, 'W')
pm10_a22_smooth = pm10_a22.Valore.ewm(span=30*24).mean()

fig = px.line(pm10_a22_resampled, x="Data", y="Valore", markers=True)
fig.show()
```
